```python
# import numpy as np  # Used to create result 3D array
from dateutil.parser import parse # Used to parse time strings into datetime objects
from math import sqrt # Used to calculate standard deviation
import matplotlib.pyplot as plt # Used for plotting data
from numpy import random

# invoke matplotlib magic command
%matplotlib inline 

MAX_SCENARIO = 4 # Used to go through scenarios, change to 1 for debugging (only go through local machine scenario
print("Number of scenarios analyzed: " + str(MAX_SCENARIO) +'\n')

# Array for retreiving and storing times
receivedTimes = [[],[]] # receivedTime [[clientReceivedTimes], [serverReceivedTimes]]
sentTimes = [[],[]] # sentTimes [[clientReceivedTimes], [serverReceivedTimes]]

# Lists to keep track of calculated values
latenciesList = []
averageClientList = []
averageServerList = []
totalAverageList = []
offsetList = []
delayList = []
stdDevClientList = []
stdDevServerList = []
totalStdDevList = []

scenarios = range(0, MAX_SCENARIO)
messages = range(0,5)

# Go through all scenarios times
for scenario in scenarios:
    # Go through client/server times, 0 = client, 1 = server
    for source in range(0,2):
        resultFile = "UDPServerClientOutput" + str(scenario) + str(source) + ".txt"
        with open(resultFile) as result:
            # Return a list of the lines, breaking at line broundaries, remove \n
            lines = result.read().splitlines()
            
            # Process through all retrieved content
            for resultLine in lines:
#                 print("Reading in " + str(resultLine))
                if "Host" in resultLine:
                    # Skip this line
                    continue
                if "Received" in resultLine:
                    # Increment mode (reading received or sent lines), reset resultTime counter
                    resultLine = resultLine.replace('Received: ', '')
                    receivedTimes[source].append(parse(resultLine))
#                     print("Read in as " + str(parse(resultLine)))
                elif "Sent" in resultLine:
                    resultLine = resultLine.replace('Sent: ', '')
                    sentTimes[source].append(parse(resultLine))
#                     print("Read in as " + str(parse(resultLine)))
    # At this point, for this scenario, the time lists should hold all the generated result times 
    # Now, first we'll want to start anaylzying the latencies/statistics

    # Q1:
    # Compute the pairwise latencies (A to B and B to A) or (T2 - T1) and (T4 - T3)
    # where
    # T1 = client timestamp on send message time, T2 = server timestamp on receiving message time,
    # T3 = server timestamp on reply message time, T4 = client timestamp on receiving reply time
    # 
    # ____ T2 ____T3 ____ Server
    #     /        \
    # __T1 ________T4 ___ Client
    #
    # For our lists:
    # T1 = sentTimes[client][message]
    # T2 = receivedTimes[server][message]
    # T3 = sentTimes[server][message]
    # T4 = receivedTimes[client][message]
    # where client = 0, server = 1
    
    # Append spots for new scenario data in lists
    latenciesList.append([])
    latenciesList[scenario].append([]) # Track latencies from client -> server
    latenciesList[scenario].append([]) # Track latencies from server -> client
    offsetList.append([])
    delayList.append([])
    
    for message in messages:
        client = 0 # define here for ease of reading during calculations
        server = 1
        
        # Read time client sent message
        T1 = sentTimes[client][message]
        # Read time client sent message
        T2 = receivedTimes[server][message]
        
        # Read time server sent reply
        T3 = sentTimes[server][message]
        # Read time client received reply
        T4 = receivedTimes[client][message]
        
        # Calculate sent latency, client -> server
        deltaA = T2 - T1
        # Convert to total microseconds difference
        deltaA = deltaA.total_seconds()*1000000
        
        # Calculate reply latency, server -> client
        deltaB = T4 - T3
        # Convert to total microseconds difference
        deltaB = deltaB.total_seconds()*1000000
        
        # Add to latencies list
        latenciesList[scenario][client].append(deltaA)
        latenciesList[scenario][server].append(deltaB)
        
        # Compute the offset and delay for each of the five time measurements using the NTP formula, as shown below:
        # offset = (T2 - T1 - T4 + T3)/2 = [(T2 - T1) - (T4 - T3)]/2 = (deltaA - deltaB)/2
        offset = (deltaA - deltaB)/2
        offsetList[scenario].append(offset)

        # delay = T2 - T1 + T4 - T3 = (T2 - T1) + (T4 - T3) = deltaA + deltaB
        # Add to the list of delays
        delay = deltaA + deltaB
        delayList[scenario].append(delay)
        
    # Compute average and standard deviations among latencies
    # Average:
    totalAverage = 0 # Average all 10 message latency pairs
    
    for direction in range (0,2):# direction = 0: client -> server, = 1: server -> client
        average = 0
        for message in messages:
            # Add together all the latencies
            average += latenciesList[scenario][direction][message]
        totalAverage += average
        average /= 5
        # Add to the averageList
        if (direction == 0):
            averageClientList.append(average)
        else:
            averageServerList.append(average)
    totalAverage /= 10
    # Add to the totalAverageList
    totalAverageList.append(totalAverage)

    # Standard deviation: for each latency, subtract the mean and square the result
    # Then work out the mean of those squared differences
    # Then take the square root of that mean to calculate standard deviation
    stdDev = 0
    totalStdDev = 0
    for direction in range (0,2):
        latency = 0
        for message in messages:
            latency = latenciesList[scenario][client][message]
            latency -= average # Subtract the mean
            latency *= latency # Square the result
            stdDev += latency  # Find new mean
        totalStdDev += stdDev
        stdDev /= 5
        stdDev = sqrt(stdDev) # Take square root of new mean
        # Add to the averageList
        if (direction == 0):
            stdDevClientList.append(stdDev)
        else:
            stdDevServerList.append(stdDev)
    totalStdDev /= 10
    totalStdDev = sqrt(totalStdDev)
    totalStdDevList.append(totalStdDev)

    # At this point, the calculations should all be done and the next step is to plot to facilitate analysis    
    # Plot each message's calculations
    # Latencies
    plt.figure()
    plt.plot(messages,latenciesList[scenario][client],'bo-', label=' Message') 
    plt.plot(messages,latenciesList[scenario][server],'r^--', label = 'Reply')
    plt.title('Message Latencies: Scenario ' + str(scenario))
    plt.ylabel('Time for Message Transmission [us]')
    plt.xlabel('Message Number')
    plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.0,0.5))
    
    # Delays
    plt.figure()
    plt.plot(messages, delayList[scenario],'g^-', label='Delay')
    plt.title('Message Delays: Scenario ' + str(scenario))
    plt.ylabel('Delay [us]')
    plt.xlabel('Message Number')
    plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.0,0.5))
    
    # Offsets
    plt.figure()
    plt.plot(messages, offsetList[scenario],'c*--', label='Offset')
    plt.title('Message Offsets: Scenario ' + str(scenario))
    plt.xlabel('Message Number')
    plt.ylabel('Offset [us]')
    plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.0,0.5))
    
# All calculations for all scenarios should have been accomplished by now, print final results
print('latenciesList = ' + str(latenciesList))
print('averageClientList = ' + str(averageClientList))
print('averageServerList = ' + str(averageServerList))
print('totalAverageList = ' + str(totalAverageList))
print('offsetList = ' + str(offsetList))
print('delayList = ' + str(delayList))
print('stdDevClientList = ' + str(stdDevClientList))
print('stdDevServerList = ' + str(stdDevServerList))
print('totalStdDevList = ' + str(totalStdDevList))

# Plot average latencies
plt.figure()
plt.plot(scenarios, averageClientList,'bo', label='Message')
plt.plot(scenarios, averageServerList,'r^', label ='Reply')
plt.plot(scenarios, totalAverageList, 'g+--', label ='Combined')
plt.title('Average Message Latencies')
plt.ylabel('Average Time for Message Transmission [us]')
plt.xlabel('Scenario Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.0,0.5))
    
# Plot standard deviations
plt.figure()
plt.plot(scenarios, stdDevClientList,'bo', label='Message')
plt.plot(scenarios, stdDevServerList,'r^', label ='Reply')
plt.plot(scenarios, totalStdDevList, 'g+--', label = 'Combined')
plt.title('Standard Deviation in Message Latencies')
plt.ylabel('Average Time for Message Transmission [us]')
plt.xlabel('Scenario Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.,0.5))

markers = ['*', '+', '^', 'o']

# All Latencies
plt.figure()
for scenario in scenarios:
    plt.plot(messages,latenciesList[scenario][client],color = random.rand(3,1), marker = markers[scenario], label='Scenario ' + str(scenario)) 
plt.title('Message Latencies for All Scenarios: Sending Message')
plt.ylabel('Time for Message Transmission [us]')
plt.xlabel('Message Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.,0.5))

plt.figure()
for scenario in scenarios:
    plt.plot(messages,latenciesList[scenario][server],color = random.rand(3,1), marker = markers[scenario], label='Scenario ' + str(scenario)) 
plt.title('Message Latencies for All Scenarios: Sending Reply')
plt.ylabel('Time for Message Transmission [us]')
plt.xlabel('Message Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.,0.5))

# All Offsets
plt.figure()
for scenario in scenarios:
    plt.plot(messages,offsetList[scenario],color = random.rand(4,1), marker = markers[scenario], label='Scenario ' + str(scenario)) 
plt.title('Calculated Offsets for All Scenarios')
plt.ylabel('Offset [us]')
plt.xlabel('Message Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.,0.5))

# All Delays
plt.figure()
for scenario in scenarios:
    plt.plot(messages,delayList[scenario],color = random.rand(3,1), marker = markers[scenario], label='Scenario ' + str(scenario)) 
plt.title('Calculated Delays for All Scenarios')
plt.ylabel('Delay [us]')
plt.xlabel('Message Number')
plt.legend(loc = 'center left', shadow = True, bbox_to_anchor=(1.,0.5))
```
